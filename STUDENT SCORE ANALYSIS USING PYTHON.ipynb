```python
#change weekly study hours column
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
#gender distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analyzed that:
#the number of females in the data is more than the number of males
```

```python
#effect of parents on scores
gp=df.groupby("ParentEduc").agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
print(gp)
```

```python
sns.heatmap(gp,annot=True)
plt.title("Relationship between Parents Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of parents have a good impact
```

```python
gp2=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
print(gp2)
```

```python
sns.heatmap(gp2,annot=True)
plt.title("Relationship between Parents Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that parents marital status has no impact on children
```
